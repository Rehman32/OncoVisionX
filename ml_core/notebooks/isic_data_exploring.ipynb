```python
import pandas as pd

```

```python
# 1. Load your current HAM10000 data
ham_df = pd.read_csv('../data/raw/HAM10000_metadata')

# 2. Load ISIC 2019 data
isic_df = pd.read_csv('../data/raw/ISIC_2019_Training_Metadata.csv')

```

```python
# 3. Find unique images in ISIC 2019 that are NOT in HAM10000
# HAM10000 images are all inside ISIC 2019, so we filter them out
isic_unique_df = isic_df[~isic_df['image'].isin(ham_df['image_id'])]

print(f"Total ISIC 2019 images: {len(isic_df)}")
print(f"Duplicates (HAM10000): {len(isic_df) - len(isic_unique_df)}")
print(f"New unique images to add: {len(isic_unique_df)}")
```
