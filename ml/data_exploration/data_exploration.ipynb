```python
import os
BASE_PATH = r"E:\CancerVision_Data\data"

DATA_PATHS = {
    # Radiology
    "radiology": os.path.join(BASE_PATH, "radiology", "TCGA-LUAD"),

    # Clinical
    "clinical": os.path.join(BASE_PATH, "Clinical", "Clinical_data"),

    # Genomics
    "genomics": {
        "mutations": os.path.join(BASE_PATH, "Genomics", "Mutations", "mutations"),
        "rnaseq": os.path.join(BASE_PATH, "Genomics", "RNASeq", "RNAseq"),
    },

    # Pathology (multiple physical locations)
    "pathology": [
        os.path.join(BASE_PATH, "pathology", "pathology_1"),
        os.path.join(BASE_PATH, "pathology", "pathology_2"),
        os.path.join(BASE_PATH, "pathology", "pathology_3"),
        r"F:\FYP_Preparation\Data\pathalogy\pathology_4",
    ],
}
```

```python
def explore_directory(path, max_items=5):
    if not os.path.exists(path):
        return {"path": path, "exists": False}

    items = os.listdir(path)
    return {
        "path": path,
        "exists": True,
        "num_items": len(items),
        "sample_items": items[:max_items],
    }
```

```python
print("\n=== DATASET STRUCTURE OVERVIEW ===\n")

# Radiology
print("Radiology:")
print(explore_directory(DATA_PATHS["radiology"]))

# Clinical
print("\nClinical:")
print(explore_directory(DATA_PATHS["clinical"]))

# Genomics
print("\nGenomics:")
for g_type, g_path in DATA_PATHS["genomics"].items():
    print(f"  {g_type}:")
    print("   ", explore_directory(g_path))

# Pathology (multiple roots)
print("\nPathology:")
for idx, p_path in enumerate(DATA_PATHS["pathology"], start=1):
    print(f"  pathology_{idx}:")
    print("   ", explore_directory(p_path))
```
