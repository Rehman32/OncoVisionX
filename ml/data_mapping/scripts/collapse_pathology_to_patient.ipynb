```python
import pandas as pd
from pathlib import Path
```

```python

INPUT_FILE = Path("../processed/pathology_file_to_case.csv")
OUTPUT_FILE = Path("../processed/pathology_cases_patient.csv")

def main():
    df = pd.read_csv(INPUT_FILE)
    print(f"Total file-level rows: {len(df)}")

    # Collapse to patient
    df_patient = df.groupby("case_id").agg(
        has_pathology=("local_verified", "max"),
        num_pathology_files=("file_id", "count")
    ).reset_index()

    print(f"Unique patients with pathology: {len(df_patient)}")

    df_patient.to_csv(OUTPUT_FILE, index=False)
    print(f"Saved patient-level pathology mapping: {OUTPUT_FILE.resolve()}")

if __name__ == "__main__":
    main()
```
