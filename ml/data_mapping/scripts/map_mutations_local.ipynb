```python
import pandas as pd
from pathlib import Path
```

```python

SAMPLE_SHEET = Path(r"E:\CancerVision_Data\data\Genomics\Mutations\mutation-sample_sheet.tsv")
LOCAL_DIR = Path(r"E:\CancerVision_Data\data\Genomics\Mutations\mutations")
OUTPUT_FILE = Path("../processed/genomics_mutations_file_to_case.csv")

def main():
    df = pd.read_csv(SAMPLE_SHEET, sep="\t")
    df.columns = [c.lower().strip().replace(" ", "_") for c in df.columns]

    local_ids = {p.name for p in LOCAL_DIR.iterdir() if p.is_dir()}
    print(f"Local mutation folders: {len(local_ids)}")

    df["local_verified"] = df["file_id"].isin(local_ids).astype(int)
    df_local = df[df["local_verified"] == 1]

    print(f"Matched mutation files: {len(df_local)}")

    df_local = df_local[[
        "file_id", "case_id", "sample_id", "tissue_type",
        "specimen_type", "local_verified"
    ]]

    df_local.to_csv(OUTPUT_FILE, index=False)
    print(f"Saved: {OUTPUT_FILE.resolve()}")

if __name__ == "__main__":
    main()
```
